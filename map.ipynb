```python
import folium
import pandas as pd
```

```python
test_cities = pd.read_csv('test_city.csv')
```

```python
test_cities.describe()
```

```python
map = folium.Map(location=[35.832660,39.726940], zoom_start=3)
 
for i, r in test_cities.iterrows():
    folium.Marker(location=[r['lat'], r['lon']], popup=r['City']).add_to(map)
```
